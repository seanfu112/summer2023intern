```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df1 = df.pivot_table(index="busi_date",columns="industry", values = "mktval")
```

```python
df1
```

```python
df1.index = pd.to_datetime(df1.index, format = "%Y/%m/%d")
```

```python
df1 = df1.sort_index()
```

```python
df1.fillna(value=0)
```

```python
df1 = df1.T
```

```python
index = df1.index.tolist()
replaced = []
for x in index:
    temp = x.replace("ind","")
    temp = temp.zfill(2)
    temp = f"ind{temp}"
    replaced.append(temp)
df1.index = replaced
df1 = df1.sort_index()
```

```python
df1
```

```python
df1 = df1.T
```

```python
import matplotlib.pyplot as plt
```

```python
df1
```

```python
plt.figure()
```

```python
df1.plot(figsize = (10,9))
plt.legend(ncol=4, loc="best")
plt.show()
```
