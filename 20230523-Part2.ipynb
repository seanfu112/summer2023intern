```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df_div = pd.read_csv("data20230523.csv")
```

```python
df_div
```

```python
df1 = df.pivot_table(index="busi_date",columns="wind_stkcode", values = "daily_profit")
```

```python
df1
```

```python
df_div = df_div.pivot_table(index="busi_date", values = "prev_ttl_asset")
```

```python
df1.index = pd.to_datetime(df1.index, format = "%Y/%m/%d")
```

```python
df_div.index = pd.to_datetime(df_div.index, format = "%Y/%m/%d")
```

```python
df1.index = df1.index.astype(str)
```

```python
df_div.index = df_div.index.astype(str)
```

```python
df1 = df1.sort_index()
```

```python
df_div = df_div.sort_index()
```

```python
df1 = pd.DataFrame(df1.sum(axis=1), index = df1.index.tolist(), columns = ["total_profit"])
```

```python
df_final = df1.join(df_div)
```

```python
df_final
```

```python
df_final["Results"] = (df_final["total_profit"]/df_final["prev_ttl_asset"])
```

```python
df_final
```

```python
df_final = df_final.pivot_table(index=df_final.index,values="Results")
```

```python
import matplotlib.ticker as mtick
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
colors = ["green" if value < 0 else "red" for value in df_final["Results"]]
plt.figure(figsize = (10,9))
plt.bar(df_final.index,df_final["Results"],color=colors)
plt.title("日收益率")
plt.xlabel("日期")
plt.ylabel("收益率")
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.show()
```

```python
df_final
```

```python
results = df_final.apply(lambda x: np.expm1(np.log1p(x).cumsum()))
```

```python
results
```

```python
plt.figure(figsize= (20,10))
plt.subplot(1,2,1)
colors = ["green" if value < 0 else "red" for value in df_final["Results"]]
plt.bar(df_final.index,df_final["Results"],color=colors)
plt.title("日收益率图")
plt.xlabel("日期")
plt.ylabel("收益率")
plt.xticks(rotation = 45)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.subplot(1,2,2)
plt.plot(results.index,results["Results"])
plt.title("累计收益率图")
plt.xlabel("日期")
plt.ylabel("收益率")
plt.xticks(rotation = 45)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.show()
```
